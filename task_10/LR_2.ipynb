```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Рандомно задаем точки
count_of_dots = 100
```

```python
# coordinates (2d)
x = np.random.randint(0, 100, count_of_dots)
y = np.random.randint(0, 100, count_of_dots)
```

```python
points = []
for i in range(count_of_dots):
    points.append([x[i], y[i]])
```

```python
frame = pd.DataFrame(points)
X = frame.iloc[:, 0]
Y = frame.iloc[:, 1]
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
m = 0
c = 0
L = 0.0001
epochs = 1000
n = float(len(X))

# градиентный спуск
for i in range(epochs):
    Y_pred = m * X + c  # Предсказанное значение
    D_m = (-2 / n) * sum(X * (Y - Y_pred))  # Производная wrt m
    D_c = (-2 / n) * sum(Y - Y_pred)  # Производная wrt c
    m = m - L * D_m  # обновляем m
    c = c - L * D_c  # обновляем c

# Предсказание
Y_pred = m * X + c
```

```python
plt.scatter(X, Y)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # линия регрессии
plt.show()
```
