```python
# Prim's Minimum Spanning Tree (MST) algorithm
# (с матричным представлением графа)
 
import sys # нужно для INT_MAX (maxsize)
```

```python
class Graph():
 
    def __init__(self, vertices):
        self.V = vertices
        self.graph = [[0 for column in range(vertices)] 
                    for row in range(vertices)]
 
    # для печати созданного MST, хранящегося в parent[]
    def printMST(self, parent):
        print("Ребро \tВес")
        for i in range(1, self.V):
            print(parent[i], "-", i, "\t", self.graph[i][ parent[i] ])
 
    # для поиска вершины с минимальным значением расстояния из набора вершин, 
    # еще не включенных в дерево кратчайших путей.
    def minKey(self, key, mstSet):
        # Инициализировать минимальное значение
        min = sys.maxsize
        for v in range(self.V):
            if key[v] < min and mstSet[v] == False:
                min = key[v]
                min_index = v
        return min_index
 
    # Функция для построения и печати MST для графа, представленного с использованием представления матрицы смежности
    def primMST(self):
        # Key values, используемые для выбора минимального веса при резке
        key = [sys.maxsize] * self.V
        parent = [None] * self.V # Массив для хранения созданного MST
        # Сделать ключ 0 так, чтобы эта вершина была выбрана первой вершиной
        key[0] = 0
        mstSet = [False] * self.V
        parent[0] = -1 # Первый узел всегда является корнем
        for cout in range(self.V):
 
            # Выберите вершину с минимальным расстоянием из набора еще не обработанных вершин. u всегда равно src на первой итерации
            u = self.minKey(key, mstSet)
 
            # Поместите вершину с минимальным расстоянием в дерево кратчайших путей
            mstSet[u] = True
 
            # Обновлять значение расстояния для смежных вершин выбранной вершины, 
            # только если текущее расстояние больше нового расстояния, а вершина не входит в дерево путей для пробного теста
            for v in range(self.V):
 
                # graph[u][v] отличен от нуля только для смежных вершин из m. mstSet[v] ложен для вершин, 
                # еще не включенных в MST. Обновить ключ, только если graph[u][v] меньше, чем key[v]
                if self.graph[u][v] > 0 and mstSet[v] == False and key[v] > self.graph[u][v]:
                        key[v] = self.graph[u][v]
                        parent[v] = u
 
        self.printMST(parent)
```

```python
g = Graph(5)
g.graph = [ [0, 2, 0, 6, 0],
            [2, 0, 3, 8, 5],
            [0, 3, 0, 0, 7],
            [6, 8, 0, 0, 9],
            [0, 5, 7, 9, 0]]
```
